```python
import gdsfactory as gf
from ADF_PhC import ADF_PhC

c = gf.Component()

D = ADF_PhC(
    A            = [0.010, 0.010, 0.010, 0.010,0.050, 0.050, 0.080, 0.150],
    M            = [319, 320, 324, 325,326, 327, 331, 332],
    WgWidth      = 0.6,
    WgWidthIO    = 0.6,
    BendRadius   = 30,
    BendRadiusIO = 20,
    Gap          = 0.2,
    FAGap        = 127,
    InIO         = 2,
    OutIO        = 2,
    OutputIO     = True,
    TaperOn      = True,
    BufLength    = 25,
)

c << D
c.plot()

c.write_gds("ADF_PhC_ADF31_test.gds")
print("Written ADF_PhC_ADF31_test.gds")
```
