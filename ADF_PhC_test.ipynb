```python
#---------------------------------------------------------------------------
# Imports
#---------------------------------------------------------------------------
import gdsfactory as gf
from ADF_PhC import ADF_PhC

#---------------------------------------------------------------------------
# Build Device
#---------------------------------------------------------------------------
C = gf.Component()


D = ADF_PhC(
    A            = [0.010,0.015],
    M            = [322,318],
    BendRadius   = 30.0,
    BendRadiusIO = 20.0,
    WgWidth      = 0.6,
    WgWidthIO    = 0.6,
    Gap          = 0.150,
    FAGap        = 50,        
    InIO         = 1,
    OutIO        = 4,
    OutputIO     = False,
    TaperOn      = True,
    BufLength    = 25,
    NPoints      = 360*322
)

C << D

#---------------------------------------------------------------------------
# Show / Save
#---------------------------------------------------------------------------
C.plot()
C.write_gds("APF30_PhC_M322_M318_A10nm_A15nm_test.gds")
```
